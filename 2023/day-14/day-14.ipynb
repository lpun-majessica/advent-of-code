```python
from pathlib import Path
import numpy as np
```

```python
with Path("./day-14.txt").open() as file:
    position = np.array([list(x.strip("\n")) for x in file.readlines()])
```

# Total load on the north support beams

## Part 1

```python
sum_1 = 0

for col in position.T:
    load = position.shape[0]
    for index, pos in enumerate(col, 1):
        if pos == "#":
            load = position.shape[0] - index
        if pos == "O":
            sum_1 += load
            load -= 1

sum_1
```
