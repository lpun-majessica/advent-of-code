# Import *Map ~~of the day~~*

```python
from pathlib import Path
import numpy as np
```

```python
with Path("./day-21.txt").open() as file:
    map_ = np.array([list(x.strip("\n")) for x in file.readlines()])
```

# Total reachable *Garden plots*

## Part 1

```python
rocks = []
starts = []

for coord, tile in np.ndenumerate(map_):
    if tile == "#":
        rocks.append(coord)
    elif tile == "S":
        starts.append(coord)
```

```python
for i in range(64):
    next = []

    for start in starts:
        next_tiles = [
            (start[0] + 1, start[1]),
            (start[0] - 1, start[1]),
            (start[0], start[1] + 1),
            (start[0], start[1] - 1),
        ]

        for next_tile in next_tiles:
            if next_tile not in rocks:
                next.append(next_tile)

    starts = set(next)

len(starts)
```
